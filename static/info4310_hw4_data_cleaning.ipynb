```python
import numpy as np
import pandas as pd
import statistics
```

```python
duba = pd.read_csv('drug-use-by-age.csv')
duba.head()
```

```python
# only want select few drugs to explore & include
# start with alc of course
alcdf = duba[["age", "alcohol_use", "alcohol_frequency"]]
alcdf = alcdf.rename(columns={"alcohol_use": "use", "alcohol_frequency":"freq"})
alcdf = alcdf.assign(drug=["alcohol"]*len(duba.index))
alcdf
```

```python
# weed
weeddf = duba[["age", "marijuana_use", "marijuana_frequency"]]
weeddf = weeddf.rename(columns={"marijuana_use": "use", "marijuana_frequency":"freq"})
weeddf = weeddf.assign(drug=["marijuana"]*len(duba.index))
weeddf
```

```python
# coke
cokedf = duba[["age", "cocaine_use", "cocaine_frequency"]]
cokedf = cokedf.rename(columns={"cocaine_use": "use", "cocaine_frequency":"freq"})
cokedf = cokedf.assign(drug=["cocaine"]*len(duba.index))
cokedf
```

```python
# hallucinogens
halldf = duba[["age", "hallucinogen_use", "hallucinogen_frequency"]]
halldf = halldf.rename(columns={"hallucinogen_use": "use", "hallucinogen_frequency":"freq"})
halldf = halldf.assign(drug=["hallucinogen"]*len(duba.index))
halldf
```

```python
# pain_releiver
prdf = duba[["age", "pain_releiver_use", "pain_releiver_frequency"]]
prdf = prdf.rename(columns={"pain_releiver_use": "use", "pain_releiver_frequency":"freq"})
prdf = prdf.assign(drug=["pain releiver"]*len(duba.index))
prdf
```

```python
out = pd.concat([alcdf, weeddf, cokedf, halldf, prdf], ignore_index=True)
out
```

```python
out.to_csv('drug_use_filtered.csv')
```
